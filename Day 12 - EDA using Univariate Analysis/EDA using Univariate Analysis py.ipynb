```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train_dataset = ("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day19-understanding-your-data-descriptive-stats/train.csv")
```

## Loading Dataset

```python
df = pd.read_csv(train_dataset)
```

```python
df.sample(5)
```

#  Categorical Data









## 1.Count Plot

```python
sns.countplot(x = "Survived", data = df)   #df["Survived"]
```

## 2.Pie Chart

```python
pie = df["Pclass"].value_counts()

plt.pie(pie.values, labels=pie.index,
        startangle=0,radius=1.2,
        shadow=True, explode=(0, 0, 0.1),
        autopct="%1.1f%%")
plt.show()
```

## 3. Barplot

```python
x =  df["Sex"].value_counts()

plt.bar(x.index, x.values)
```

# Numerical Data

## 1.Histogram

```python
plt.hist(df['Age'],bins=5)
```

## 2. Distplot

```python
sns.distplot(df['Age'])     # displot vs distplot vs histplot
```

## 3.Boxplot

```python
sns.boxplot(df['Age'])
```

```python
sns.violinplot(df['Fare'])
```
